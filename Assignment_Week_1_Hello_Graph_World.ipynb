**Krackhardt kite grapht shows the connection and relationship between individuals. For this assignment I decided to use my classmates from data620 and the teams we built. The goal is to connect the students who posted the group names on brightspace to one another and then connect the remaining students with in their group. For the purpose of this assignment I might not include every group and I recognize that with this goal the visualization might not look like a kite**

```python
import networkx as nx
import matplotlib.pyplot as plt
#NetworkX internally uses matplotlib
```

```python
G = nx.Graph() #empty gragh

nodes = ['Kossi', 'Victor', 'Garbriel', 'Will', 'Jon', 'Chafiaa', 'Puja', 'Susanna', 'Mikhail', 'Marjete']
G.add_nodes_from(nodes)  #add nodes of names to the graph
```

```python
edges = [('Kossi', 'Victor'), ('Kossi', 'Garbriel'), ('Kossi', 'Will'), ('Will', 'Jon'),
         ('Will', 'Chafiaa'), ('Will', 'Puja'), ('Puja', 'Kossi'),
         ('Puja', 'Susanna'), ('Puja', 'Mikhail'), ('Puja', 'Marjete')] #add edges to grapgh showing connection
G.add_edges_from(edges)
```

```python
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_size=1000, node_color='skyblue', font_size=10, font_weight='bold')
plt.title("Graph 620 groupmates")
plt.show() #draw graph & add labels
```

**Conclusion**: the visual isnt a perfect representation for my example but does the job of showing connections between students.
